### Testing the endpoint of our "The Office API"

#### First lets import our libraries

```python
from pymongo import MongoClient
import requests
```

Side note: You will need to be running the API fie with the command "python3 miapi.py" and have a local Mongo database installed as specified in the readme.md

#### Testing get endpoints

```python
#Get all lines for a character
url_frases = "http://localhost:5000/frases/"
person = "Jim"

frases = requests.get(url_frases + person).json()

len(frases)
```

```python
#Get episode numbers for a season
url_epis = "http://localhost:5000/season/"
season_number = "1"
episodes_season_one = requests.get(url_epis + season_number).json()
print(episodes_season_one)
```

```python
#Get characters featured in an episode
url_characters = "http://localhost:5000/characters_episode/"
episode_number = "1_1"
characters_first_epi = requests.get(url_characters + episode_number).json()
print(characters_first_epi)
```

```python
#Get all lines in an episode
url = "http://localhost:5000/lines_episode/"
param = "1_2"
lines_first_epi = requests.get(url + param).json()
print(characters_first_epi)
```

```python
#Get all lines in a season
url = "http://localhost:5000/lines_season/"
param = "1"
lines_season_one = requests.get(url + param).json()
print(lines_season_one)
```

```python
#Get average sentiment of a character
url = "http://localhost:5000/character_sentiment/"
param = "Darryl"
dwight_sentiment = requests.get(url + param).json()
print(dwight_sentiment)
```

```python
#Get average sentiment of an episode
url = "http://localhost:5000/episode_sentiment/"
param = "1_1"
sentiment_first_episode = requests.get(url + param).json()
print(sentiment_first_episode)
```

#### Testing POST endpoints

```python
#Insert a new line
toinsert = {'episode': "1_10",
 'season' : "1",
 'character_name': 'Dobby',
 'dialogue': 'Dobby is a free self'}
url = "http://localhost:5000/newline"
requests.post(url, data=toinsert)
```

```python
#Insert a new character
toinsert = {'character_name': 'Dobby',
 'lines': ["Dobby loves tesing his flask api"]}
url = "http://localhost:5000/newcharacter"
requests.post(url, data=toinsert)
```

```python
#Insert a new episode
toinsert = {'episode_id': '3_40',
 'characters': ["Dobby", "Amanda"]}
url = "http://localhost:5000/newepisode"
requests.post(url, data=toinsert)
```

```python
#Insert a new season
toinsert = {'season_num': '41',
 'episodes': ["41_1","41_2","41_3"]}
url = "http://localhost:5000/newseason"
requests.post(url, data=toinsert)
```
